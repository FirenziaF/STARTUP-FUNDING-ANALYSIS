#### Validity 
For our data to be valid, I have to verify that every column is accurate and appropriate for this analysis and remove those that are invalid.

```python
# Renamimg columns
df3.rename(columns={' funding_total_usd ':'funding_total_usd',' market ':'market'}, inplace=True)
```

#### Completeness.

To achieve completeness in our data, I will be checking for missing values in the data.

```python
# Checking for duplicates. 
def count_duplicates(data):
    duplicate_count = data.duplicated().sum()
    return duplicate_count
duplicate_count = count_duplicates(df3)
print(duplicate_count)
```

```python
#deleting duplicate rows.
df3 = df3.drop_duplicates()
print(df3.shape)
```

```python
#Checking for missing values and percentages.

missing_values = df3.isnull().sum()
percentage_missing = (missing_values / len(df3)) * 100

missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage Missing': percentage_missing})
print(missing_data)

# Plot missing values 
ax = df3.isna().sum().sort_values(ascending = False)[:15][::-1].plot(kind = 'barh', figsize = (9, 10))
plt.title('Percentage of Missing Values Per Column', fontdict={'size':15})
for p in ax.patches:
    percentage ='{:,.0f}%'.format((p.get_width()/df3.shape[0])*100)
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/2
    ax.annotate(percentage,(x,y))
```

```python
#dropping null values.
df3 = df3.dropna(how="all")

# Remove data before 1980
index_before_1980 = df3[df3['founded_year']< 1980].index
df3.drop(index_before_1980, inplace = True)
```

```python
# Type casting
df3['funding_total_usd'] = df3['funding_total_usd'].apply(str).str.replace(',','')
df3['funding_total_usd'] = df3['funding_total_usd'].replace(' -   ',0).astype("float")

# Convert founded_year column to float first to handle NaN values
df3['founded_year'] = df3['founded_year'].astype(float)
df3['funding_rounds'] = df3['funding_rounds'].astype(int)
df3['first_funding_at'] = pd.to_datetime(df3['first_funding_at'], errors='coerce')
df3['last_funding_at'] = pd.to_datetime(df3['last_funding_at'], errors='coerce')

# Remove NaN values
df3.dropna(inplace=True)

# Convert founded_year column to integer after removing NaN values
df3['founded_year'] = df3['founded_year'].astype(int)

# Remove data with zero funds
index_no_funding = df3[df3['funding_total_usd'] <= 0].index
df3.drop(index_no_funding, inplace=True)
```

```python
#Confirming for missing values.
has_missing_values = df3.isnull().any().any()
print(has_missing_values)
```

#### Dealing with outliers: 

Dealing with the extreme or unusual data points that can adversely impact the model's performance. 

```python
# Visualizing outliers. 

# Create box plots for numerical columns
plt.figure(figsize=(10, 8))
sns.boxplot(data=df3)
plt.xticks(rotation=90)
plt.show()
```

```python
# Convert DataFrame to numeric type
df3_numeric = df3.apply(pd.to_numeric, errors='coerce')

# Calculate the upper and lower thresholds for outlier detection
upper_threshold = 3 * df3_numeric.std() + df3_numeric.mean()
lower_threshold = df3_numeric.mean() - 3 * df3_numeric.std()

# Replace the outliers with the column's median
df3_numeric = df3_numeric.apply(lambda x: x.where(~((x > upper_threshold[x.name]) | (x < lower_threshold[x.name]))), axis=0)

# Check if any outliers are still present
outliers_present = (df3_numeric > upper_threshold) | (df3_numeric < lower_threshold)

# Return True if any outliers are found, False otherwise
data_contains_outliers = outliers_present.any().any()
print(data_contains_outliers)
```
